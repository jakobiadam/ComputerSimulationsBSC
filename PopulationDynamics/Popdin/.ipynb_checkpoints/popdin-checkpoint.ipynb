# Imports

```python
import numpy as np
import matplotlib.pyplot as plt
import math
import os
import subprocess
import cv2
import glob
from scipy.optimize import curve_fit
import seaborn as sns

from IPython.display import clear_output

#plot beállítások és paraméterek
sns.set_theme()
style = {'xtick.labelsize' : 15, 'ytick.labelsize' : 15}
plt.style.use(style)

t_s = 30
xy_s = 25
hs = 0.5
ws = 0.5
```

# 1. Feladat

**Nézzük meg, hogy a fixpontok közelében és azoktól távol a logisztikus egyenlet analitikusan kiszámolható viselkedését milyen jól adja vissza az Euler-módszer és az adaptív Runge-Kutta módszer. Ábrázoljuk együtt a numerikus és analitikus megoldásokat, vizsgáljuk meg a különbségüket! Különböző kezdő- és különböző r paraméterekkel indítva a logisztikus egyenletet rekonstruáljuk a fólia ábráit!**

## Forráskód futtatása, fájlok legyártása

```python
# params: r, x0, T, dt, accuracy, file_path1, file_path2
src = "./f1.bin"
r = [-1, -0.5, 0, 0.5, 1]
r_string = ["r-1","r-0.5","r0","r0.5","r1"]
x0 = np.round_(np.linspace(0,1,11), decimals = 3)
T = 10
dt = 0.1
accuracy = 0.01

for i in range(len(r)):
    for j in range(len(x0)):
        file_path1 = f"./Feladatok/f1/{r_string[i]}/EA{x0[j]}.txt" #Euler + Analitikus
        file_path2 = f"./Feladatok/f1/{r_string[i]}/R{x0[j]}.txt" #Runge-Kutta (adaptív negyedrendű)
        subprocess.run([src,f"{r[i]}",f"{x0[j]}",f"{T}",f"{dt}",f"{accuracy}",f"{file_path1}",f"{file_path2}"])
```

## Ábrák elkészítése és mentése

```python
fig, ax = plt.subplots(5,1,figsize=(15,42))

for i in range(len(r)):
    ax[i].plot([],[],'.',color="blue",label="Euler")
    ax[i].plot([],[],'--',color="red",label="Analitic")
    ax[i].legend(fontsize=0.9*xy_s)
    for j in range(len(x0)):
        ea = np.genfromtxt(f"./Feladatok/f1/{r_string[i]}/EA{x0[j]}.txt",dtype=float).T
        ax[i].plot(ea[0],ea[1],'.',color="blue")
        ax[i].plot(ea[0],ea[2],'--',color="red")
        ax[i].set_xlabel("t",fontsize=xy_s)
        ax[i].set_ylabel("x",fontsize=xy_s)
        ax[i].set_title(f"r={r_string[i][1:]}",fontsize=t_s)
plt.savefig("./imgs/f1/EA.jpg")
```

```python
fig, ax = plt.subplots(5,1,figsize=(15,42))

for i in range(len(r)):
    ax[i].plot([],[],'.',color="green",label="ARK4")
    ax[i].plot([],[],'--',color="red",label="Analitic")
    ax[i].legend(fontsize=0.9*xy_s)
    for j in range(len(x0)):
        ea = np.genfromtxt(f"./Feladatok/f1/{r_string[i]}/EA{x0[j]}.txt",dtype=float).T
        R = np.genfromtxt(f"./Feladatok/f1/{r_string[i]}/R{x0[j]}.txt",dtype=float).T
        ax[i].plot(R[0],R[1],'.',color="green")
        ax[i].plot(ea[0],ea[2],'--',color="red")
        ax[i].set_xlabel("t",fontsize=xy_s)
        ax[i].set_ylabel("x",fontsize=xy_s)
        ax[i].set_title(f"r={r_string[i][1:]}",fontsize=t_s)
plt.savefig("./imgs/f1/R.jpg")
```

Az adaptív Runge-Kutta módszer segítségével mindössze pár pont felhasználásával is nagyon pontos eredményt kaphatunk, szemben az Euler-módszerrel, ahol az állandó lépéshossz miatt pontjaink száma sokszorosa a Runge-Kutta módszer által használt mennyiségnek, az eredményünk mégis pontatlanabb. A jelenlegi felbontásban ez így is alig észrevehető. Megfigyelhető még, hogy minél távolabb vagyunk a fix ponttól, az eltérés annál nagyobb. Ennek oka valószínűleg az, hogy a fix pont közelében kisebbek a deriváltak, így ott a hiba is kissebb.

# 2. Feladat

**Készítsük el a fajok közös erőforrásokért való versengését modellező csatolt logisztikus modellt! Szimuláljuk az $\alpha = \beta = 1$ esetet, és demonstráljuk numerikusan a kompetitív kizárás törvényét: azaz a két faj nem létezhet stabilan együtt - a nagyobb $k$ értékű kiszorítja a másikat. Mutassuk meg, hogy a két faj együttélése csak $\alpha k_2 < k_1$ és $\beta k_1 < k_2$ esetében stabil. Tipp: szimuláljunk különböző paraméterekkel, és ábrázoljuk a hosszabb idő után megmaradt fajok számát az $(\alpha k_2 - k_1)-(\beta k_1 - k_2)$ síkon.**

Ennél a feladatnál már kizárólagosan csak az adaptív negyedrendű Runge-Kutta módszert tervezem használni.

## Kompetitív kizárás tövénye ($\alpha = \beta = 1$)

### Forráskód futtatása, fájlok legyártása
Két esetet vizsgálok, $k1 < k2$ és $k1 > k2$, minden más paraméter előre kiválasztom és változatlanul hagyom.

```python
# params: r1, r2, n1_start, n2_start, alpha, beta, k1, k2, T, dt, accuracy, file_path
src = "./f2.bin"
r1 = 1
r2 = 1
n1_start = 2000
n2_start = 1000
alpha = 1
beta = 1
k1 = 5000
k2 = 10000
T = 20
dt = 0.1
accuracy = 0.000001
file_path = "./Feladatok/f2/ab1.txt"

subprocess.run([src,f"{r1}",f"{r2}",f"{n1_start}",f"{n2_start}",f"{alpha}",f"{beta}",f"{k1}",
                f"{k2}",f"{T}",f"{dt}",f"{accuracy}",f"{file_path}"])



k1 = 4000
k2 = 3000
file_path = "./Feladatok/f2/ab2.txt"

subprocess.run([src,f"{r1}",f"{r2}",f"{n1_start}",f"{n2_start}",f"{alpha}",f"{beta}",f"{k1}",
                f"{k2}",f"{T}",f"{dt}",f"{accuracy}",f"{file_path}"])
```

### Ábrák elkészítése és mentése

```python
arr1 = np.genfromtxt("./Feladatok/f2/ab1.txt",dtype=float).T
arr2 = np.genfromtxt("./Feladatok/f2/ab2.txt",dtype=float).T

fig, axes = plt.subplots(1,2,figsize = (24,12))

axes[0].plot(arr1[0],arr1[1],'--',linewidth=4,label = "$faj_1$")
axes[0].plot(arr1[0],arr1[2],'--',linewidth=4,label = "$faj_2$")
axes[0].set_xlabel("t",fontsize=30)
axes[0].set_ylabel("Egyedszám",fontsize=30)
axes[0].set_title("Kompetitív kizárás ($k_1 < k_2$)",fontsize=40)
axes[0].legend(fontsize=24)

axes[1].plot(arr2[0],arr2[1],'--',linewidth=4,label = "$faj_1$")
axes[1].plot(arr2[0],arr2[2],'--',linewidth=4,label = "$faj_2$")
axes[1].set_xlabel("t",fontsize=30)
axes[1].set_ylabel("Egyedszám",fontsize=30)
axes[1].set_title("Kompetitív kizárás ($k_1 > k_2$)",fontsize=40)
axes[1].legend(fontsize=24)

plt.savefig("./imgs/f2/kizaras.jpg")
```

A kompetitív kizárás törvénye a fenti ábrán megfigyelhető, a kisebb kapacitású faj alulmarad a nagyobb kapacitású fajjal szemben.

## Két faj együttélése ($\alpha k_2 < k_1$ és $\beta k_1 < k_2$)

### Forráskód futtatása, fájlok legyártása
Itt négy szükséges paraméterem van, $\alpha$, $\beta$, $k_1$ és $k_2$. Ebből a kapacitások értékeit előre eldöntöttem, és csak az $\alpha$ és $\beta$ értékeket változtattam.

```python
# params: r1, r2, n1_start, n2_start, alpha, beta, k1, k2, T, dt, accuracy, file_path
src = "./f2.bin"
r1 = 1
r2 = 1
n1_start = 1000
n2_start = 100
k1 = 2000
k2 = 5000
T = 1000
dt = 0.1
accuracy = 0.000001

alpha = np.round_(np.linspace(0,3,51),decimals = 1)
beta = np.round_(np.linspace(0,3,51),decimals = 3)

for i in range(len(alpha)):
    for j in range(len(beta)):
        file_path = f"./Feladatok/f2/ketfaj/{i*len(beta) + j}.txt"
        subprocess.run([src,f"{r1}",f"{r2}",f"{n1_start}",f"{n2_start}",f"{alpha[i]}",f"{beta[j]}",f"{k1}",
                f"{k2}",f"{T}",f"{dt}",f"{accuracy}",f"{file_path}"])
        clear_output()
        print(f"{i*len(beta) + j + 1} / {len(alpha)*len(beta)}")
```

### Fájlok beolvasása, ábrák elkészítése és mentése

```python
x = []
y = []
n1 = []
n2 = []
for i in range(len(alpha)):
    for j in range(len(beta)):
        file_path = f"./Feladatok/f2/ketfaj/{i*len(beta) + j}.txt"
        arr = np.genfromtxt(file_path,dtype=float)
        x.append(alpha[i] * k2 - k1)
        y.append(beta[j] * k1 - k2)
        n1.append(arr[-1,1])
        n2.append(arr[-1,2])
        
x = np.array(x)
y = np.array(y)
n1 = np.round_(np.array(n1), decimals=4)
n2 = np.round_(np.array(n2), decimals=4)
n1n2 = np.round_(n1 * n2, decimals=4)
```

```python
fig, ax = plt.subplots(3,1,figsize=(13,28))
n_min = min(n1)
n_max = max(n1)
cs = ax[0].scatter(x,y,s=35,c=n1,vmin=n_min,vmax=n_max)
cbar = fig.colorbar(cs,ax=ax[0])
cbar.ax.set_ylabel('$n_1$',fontsize = 25)
ax[0].set_xlabel(r"$\alpha k_2 - k_1$",fontsize=25)
ax[0].set_ylabel(r"$\beta k_1 - k_2$",fontsize=25)
ax[0].set_title("Első faj végső darabszám",fontsize=30)

n_min = min(n2)
n_max = max(n2)
cs = ax[1].scatter(x,y,s=35,c=n2,vmin=n_min,vmax=n_max)
cbar = fig.colorbar(cs,ax=ax[1])
cbar.ax.set_ylabel('$n_1$',fontsize = 25)
ax[1].set_xlabel(r"$\alpha k_2 - k_1$",fontsize=25)
ax[1].set_ylabel(r"$\beta k_1 - k_2$",fontsize=25)
ax[1].set_title("Második faj végső darabszám",fontsize=30)

n_min = min(n1n2)
n_max = max(n1n2)
cs = ax[2].scatter(x,y,s=35,c=n1n2,vmin=n_min,vmax=n_max)
cbar = fig.colorbar(cs,ax=ax[2])
cbar.ax.set_ylabel(r'$n_1 \cdot n_2$',fontsize = 25)
ax[2].set_xlabel(r"$\alpha k_2 - k_1$",fontsize=25)
ax[2].set_ylabel(r"$\beta k_1 - k_2$",fontsize=25)
ax[2].set_title("A két faj végső darabszámának szorzata",fontsize=30)

plt.savefig("./imgs/f2/egyutteles.jpg")
```

Látható, hogy csak $\alpha k_2 < k_1$ és $\beta k_1 < k_2$ esetén maradhat fenn egyszerre mind a két rivális faj, más esetben, kellő idő elteltével az egyik biztosan kipusztul.

# 3. Feladat

**Implementáljuk a Lotka-Volterra-modellt! Ábrázoljuk a különböző paramétereknél a fajok szaporodását és az $n_R - n_F$ fázisteret!**

## Forráskód futtatása, fájlok legyártása

Az $a$ és $d$ változókat állandónak választom, a $b$ és $c$ változók hatását vizsgálom inkább, melyek a nyulak és rókák egymásra gyakorolt hatását fejezik ki.

```python
# params: a, b, c, d, nR_start, nF_start, T, dt, accuracy, file_path
src = "./f3.bin"
a = 0.4
b = [0.001,0.001,0.004,0.004]
c = [0.001,0.004,0.001,0.004]
d = 0.9
nR_start = 500
nF_start = 200
T = 100
dt = 0.1
accuracy = 0.000001

for i in range(len(b)):
    file_path = f"./Feladatok/f3/{i}.txt"
    subprocess.run([src,f"{a}",f"{b[i]}",f"{c[i]}",f"{d}",f"{nR_start}",f"{nF_start}",
                    f"{T}",f"{dt}",f"{accuracy}",f"{file_path}"])
```

## Fájlok beolvasása, ábrák elkészítése és mentése

```python
fig, axes = plt.subplots(4,2,figsize=(20,35))

for i in range(len(b)):
    file_path = f"./Feladatok/f3/{i}.txt"
    arr = np.genfromtxt(file_path,dtype=float).T
    axes[i,0].plot(arr[0],arr[1],'--',color='blue',label="Nyulak")
    axes[i,0].plot(arr[0],arr[2],'--',color='red',label="Rókák")
    axes[i,0].set_xlabel("Idő",fontsize=25)
    axes[i,0].set_ylabel("Egyedszám",fontsize=25)
    if i == 0:
        axes[i,0].set_title(f"Szaporodási görbék\n$b={b[i]}$, $c={c[i]}$",fontsize=30)
    else:
        axes[i,0].set_title(f"$b={b[i]}$, $c={c[i]}$",fontsize=30)
    axes[i,0].legend(fontsize=22)
    
    axes[i,1].plot(arr[1],arr[2],'--',color='green')
    axes[i,1].set_xlabel("Nyulak száma",fontsize=25)
    axes[i,1].set_ylabel("Rókák száma",fontsize=25)
    if i == 0:
        axes[i,1].set_title(f"$n_R - n_F$ fázisterek\n$b={b[i]}$, $c={c[i]}$",fontsize=30)
    else:
        axes[i,1].set_title(f"$b={b[i]}$, $c={c[i]}$",fontsize=30)
    
plt.savefig("./imgs/f3/LV.jpg")
```

# 4. Feladat

**Készítsük el a véges táplálék ill. telítődés hatását is figyelembe vevő Lotka-Volterra -modellt! Ábrázoljuk itt is a populációk létszámát illetve a fázisteeret! Keressünk fixpontokat, vizsgáljuk meg azok stabilitását!**

## Forráskód futtatása, fájlok legyártása

Az $a$, $b$, $c$ és $d$ változókat állandónak választom, a $k$ és $s$ változók hatását vizsgálom inkább, melyek eltérnek a korábbi modelltől.

```python
# params: a, b, c, d, nR_start, nF_start, k, s, T, dt, accuracy, file_path
src = "./f4.bin"
a = 0.4
b = 0.004
c = 0.004
d = 0.9
nR_start = 500
nF_start = 200
k = [1e16,800,800,1000] #ha k nagyon nagy, olyan, mintha nem lenne bevezetve
s = [1000,1e16,500,270] #ha s nagyon nagy, olyan, mintha nem lenne bevezetve
T = 115
dt = 0.1
accuracy = 0.000001

for i in range(len(k)):
        file_path = f"./Feladatok/f4/{i}.txt"
        subprocess.run([src,f"{a}",f"{b}",f"{c}",f"{d}",f"{nR_start}",f"{nF_start}",f"{k[i]}",f"{s[i]}",
                        f"{T}",f"{dt}",f"{accuracy}",f"{file_path}"])
```

## Telítődés értelmezése

```python
def telFun(x,s):
    return x / (1 + x/s)

s=[1,2,5,10,50,100]
x = np.linspace(0,100,101)
fig = plt.figure(figsize=(14,10))

for i in range(len(s)):
    plt.plot(x,telFun(x,s[i]),'--',label=f"s={s[i]}")
plt.legend(fontsize=25)
plt.title("Telítődés viselkedése",fontsize=30)
plt.xlabel("Egyedszám (n)",fontsize=25)
plt.ylabel(r"$\dfrac{n}{1 + \dfrac{n}{s}}$",fontsize=25)

plt.savefig("./imgs/f4/telitodes.jpg")
```

Minél kisebb $s$ értéke, annál kevésbé tudnak a ragodozók többet fogyasztani a prédákból annak ellenére, hogy a prédák létszáma nő.

## Fájlok beolvasása, ábrák elkészítése és mentése

```python
fig, axes = plt.subplots(4,2,figsize=(20,35))

for i in range(len(k)):
    file_path = f"./Feladatok/f4/{i}.txt"
    arr = np.genfromtxt(file_path,dtype=float).T
    axes[i,0].plot(arr[0],arr[1],'--',color='blue',label="Nyulak")
    axes[i,0].plot(arr[0],arr[2],'--',color='red',label="Rókák")
    axes[i,0].set_xlabel("Idő",fontsize=25)
    axes[i,0].set_ylabel("Egyedszám",fontsize=25)
    if i == 0:
        axes[i,0].set_title(f"Szaporodási görbék\n$k={k[i]}$, $s={s[i]}$",fontsize=30)
    else:
        axes[i,0].set_title(f"$k={k[i]}$, $s={s[i]}$",fontsize=30)
    axes[i,0].legend(fontsize=22)
    
    axes[i,1].plot(arr[1],arr[2],'--',color='green')
    axes[i,1].set_xlabel("Nyulak száma",fontsize=25)
    axes[i,1].set_ylabel("Rókák száma",fontsize=25)
    if i == 0:
        axes[i,1].set_title(f"$n_R - n_F$ fázisterek\n$k={k[i]}$, $s={s[i]}$",fontsize=30)
    else:
        axes[i,1].set_title(f"$k={k[i]}$, $s={s[i]}$",fontsize=30)
    
plt.savefig("./imgs/f4/LVR.jpg")
```

Ha nem veszem számításba a nyulak véges táplálékát és a rókák telítődési száma megfelelő (vagyis nem halnak meg a rókák a nyulak kezdeti populáció létszáma mellett), akkor a nyulak és rókák populáció létszáma egymással szinkronban oszcillálva divergálni kezdenek a végtelenbe (hol felszökik a létszámuk, hol nulla közelébe esik vissza, és minden felszökés exponenciálisan növekszik az előzőekhez képest). Ha számításba veszem a nyulak véges táplálékmennyiségét és a rókák telídősési száma megfelelő, akkor a két faj tud egymás mellett létezni, mindkét faj populációs száma tart egy-egy adott értékhez. Ha a telítődési szám túl kicsi, a rókák kihalnak, a nyulak pedig beállnak a kapacitásuk értékére. Tehát a triviális (a rókák kipusztulnak) esetet leszámítva, a fixpontok a nyulak korlátozott kapacitása és a rókák megfelelően beállított telítődési száma mellett tudnak létrejönni, a lehetséges értékek közül pedig a kisebb telítődési számok mellett áll be gyorsabban az egyensúly.
