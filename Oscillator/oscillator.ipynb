```python
import numpy as np
from scipy.signal import find_peaks
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
import math
import seaborn as sns
sns.set_theme()

dic = {"xtick.labelsize" : 20, "ytick.labelsize" : 20}

plt.style.use(dic)
```

```python
xy_s = 25 #x és y label fontsize
t_s = 30 #title fontsize
hs = 0.3 #hspace
ws = 0.25 #wspace
```

```python
def beolvas(filenev):
    f = open(filenev, "r")
    lista=[i.strip().split('\t') for i in f]
    f.close()
    for i in range(len(lista)):
        for j in range(10):
            lista[i][j]=float(lista[i][j])
    return np.array(lista)
```

```python
#A
tabla=beolvas("A.txt")
plt.figure(figsize=(16,12))
plt.subplot(221)
plt.plot(tabla[:,0],tabla[:,1],'--',color='red')
plt.title("Euler-Cromer x-t",fontsize=t_s)
plt.xlabel("t",fontsize=xy_s)
plt.ylabel("x",fontsize=xy_s)
plt.subplot(222)
plt.plot(tabla[:,0],tabla[:,7],'--',color='green')
plt.title("Analitikus x-t",fontsize=t_s)
plt.xlabel("t",fontsize=xy_s)
plt.ylabel("x",fontsize=xy_s)
plt.subplot(223)
plt.plot(tabla[:,0],tabla[:,2],'--',color='red')
plt.title("Euler-Cromer v-t",fontsize=t_s)
plt.xlabel("t",fontsize=xy_s)
plt.ylabel("v",fontsize=xy_s)
plt.subplot(224)
plt.plot(tabla[:,0],tabla[:,8],'--',color='green')
plt.title("Analitikus v-t",fontsize=t_s)
plt.xlabel("t",fontsize=xy_s)
plt.ylabel("v",fontsize=xy_s)
plt.subplots_adjust(hspace=hs,wspace = ws)
plt.savefig('img/A.pdf')  
```

```python
indices=find_peaks(tabla3[:,4])

def exponent(x,a,b,c):
    return a+b*np.exp(c*x)

x=[]
y=[]
for i in indices[0]:
    x.append(tabla3[i,0])
    y.append(tabla3[i,4])

x=np.array(x)
y=np.array(y)

b=5
a=5
c=0.625
popt,pcov=curve_fit(exponent,x,y,p0=[a,b,c])
perr = np.sqrt(np.diag(pcov))
print(popt)
print(perr)
```

```python
plt.figure(figsize=(12,8))
plt.plot(tabla3[1200:,0],tabla3[1200:,4],'--',color='red')
plt.plot(tabla3[1200:,0],exponent(tabla3[1200:,0],*popt),'--',color="orange")
plt.title("Euler módszer x-t diagramjának elszállása",fontsize=27)
plt.xlabel("t",fontsize=xy_s)
plt.ylabel("x",fontsize=xy_s)
plt.savefig('img/B4.pdf')
```

```python
x=[]
for i in range(25000,500001,25000):
    x.append(i)
x=np.array(x)

f = open('t_ec.txt', "r")
y1=[i.split('\t') for i in f]
f.close()
del y1[0][-1]
y1=np.array([int(i) for i in y1[0]])

f = open('t_e.txt', "r")
y2=[i.split('\t') for i in f]
f.close()
del y2[0][-1]
y2=np.array([int(i) for i in y2[0]])

def lin(x,a,b):
    return a*x+b

a=14/5
b=0
popt1,pcov1=curve_fit(lin,x,y1,p0=[a,b])
perr1 = np.sqrt(np.diag(pcov1))
popt2,pcov2=curve_fit(lin,x,y2,p0=[a,b])
perr2 = np.sqrt(np.diag(pcov2))
```

```python
plt.figure(figsize=(12,8))
plt.plot(x,y1,"o",color='red',label="Euler-Cromer pontok")
plt.plot(x,lin(x,*popt1),color='red',label="Euler-Cromer illesztés")
plt.title("Euler-Cromer és Euler futásidő diagram",fontsize=27)
plt.xlabel("lépésszám",fontsize=xy_s)
plt.ylabel("t [μs]",fontsize=xy_s)
plt.plot(x,y2,"o",color='green',label="Euler pontok")
plt.plot(x,lin(x,*popt2),color='green',label="Euler illesztés")
plt.legend()
plt.grid(True)
plt.savefig('img/futas.pdf')  
```
